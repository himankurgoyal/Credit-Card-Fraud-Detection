```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
df = pd.read_csv("creditcard.csv")
```

```python
df.head()
```

```python
df2 = df
data = df.values
```

```python
data_x = df.drop(["Class"],axis=1)
data_y = df["Class"]
```

```python
train_x = np.array(data_x[:280000])
train_y = np.array(data_y[:280000])


test_x = np.array(data_x[280000:])
test_y = np.array(data_y[280000:])
```

```python
train_y[:5]
```

# ANN model

```python
model = Sequential()

model.add(Dense(16, activation = 'relu', input_shape=(30,)))
model.add(Dense(8, activation = 'relu'))
model.add(Dense(4, activation = 'relu'))
model.add(Dense(1))

model.compile(loss = 'mse', optimizer = "adam", metrics=['accuracy'])

```

```python
model.summary()
```

```python
hist = model.fit(train_x,train_y, validation_split = 0.2, epochs = 5)
```

```python
out = []
out = model.predict(test_x)
print(out)
```

```python
for i,j in enumerate(out):
    if j<0.5:
        out[i] = 0
    else:
        out[i] = 1 
```

```python
for i in out:
    print(i)
```

```python
def plot_learningCurve(history, epoch):
  # Plot training & validation accuracy values
  epoch_range = range(1, epoch+1)
  plt.plot(epoch_range, history.history['accuracy'])
  plt.plot(epoch_range, history.history['val_accuracy'])
  plt.title('Model accuracy')
  plt.ylabel('Accuracy')
  plt.xlabel('Epoch')
  plt.legend(['Train', 'Val'], loc='upper left')
  plt.show()

  # Plot training & validation loss values
  plt.plot(epoch_range, history.history['loss'])
  plt.plot(epoch_range, history.history['val_loss'])
  plt.title('Model loss')
  plt.ylabel('Loss')
  plt.xlabel('Epoch')
  plt.legend(['Train', 'Val'], loc='upper left')
  plt.show()
```

```python
plot_learningCurve(hist, 5)
```
